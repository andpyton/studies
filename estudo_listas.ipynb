```python
## Implementando um Bubble sort

lista1 = [0, -2, 3, 10, 22, -3]
cont = len(lista1)
for i in range (len(lista1)-1):
  cont=cont-1
  for j in range(i+1,cont):
    if lista1[j]<lista1[i]:
      prox = lista1[j]
      ant = lista1[i]
      lista1[j] = ant
      lista1[i] = prox








```

```python
# selection sort - The user chooses the size of a list and adds values ​​to it. The function will sort the list from smallest to largest value.

while True:

  try:

    size = int(input("Type size of list: "))
    break

  except ValueError:
    print("Type a correct value: ")

lista=[]

while len(lista)<size:

  try:
    var = int(input())
    lista.append(var)
  except:
    print("Type a correct value...")


def sorting_smallest_to_largest_value(lista1, size):

  i = 0

  for  i in range (size-1):

    for j in range(i+1, size):

      if lista1[j] < lista1[i]:

        bigger = lista1[i]
        minor = lista1[j]

        lista1[j] = bigger
        lista1[i] = minor

      else:
        continue

  return lista1



bubble_sort = sorting_smallest_to_largest_value(lista, size)

print(f'minor to bigger: {bubble_sort}')


# New, Pythonic swap:#
# lista1[i], lista1[j] = lista1[j], lista1[i]
```
